```python
def func_run(DF: pd.DataFrame):
    y_df = DF.iloc[:, 1]
    y_mean = np.mean(y_df)
    df_run = pd.DataFrame(columns=['Time', 'Value'])
    df_above_mean = pd.DataFrame(columns=['Time', 'Value'])
    df_below_mean = pd.DataFrame(columns=['Time', 'Value'])
    cnt_above = 0
    cnt_below = 0

    for idx, val in DF.iterrows():
        # is this is the first value in the dataframe?
        if idx==0 and val['Value']>y_mean:
            #df_above_mean = df_above_mean.append(val)
            df_above_mean = pd.concat([df_above_mean,val], ignore_index=True)
            cnt_above += 1
            cnt_below = 0
        elif idx==0 and val['Value']<=y_mean:
            #df_below_mean = df_below_mean.append(val)
            df_below_mean = pd.concat([df_below_mean,val], ignore_index=True)
            cnt_above = 0
            cnt_below += 1

        # second value
        if idx>0 and val['Value']>y_mean:
            #df_above_mean = df_above_mean.append(val)
            df_above_mean = pd.concat([df_above_mean,val], ignore_index=True)
            cnt_above += 1
            cnt_below = 0
            # check if above threshold
            if cnt_above==8:
                #df_run = df_run.append(df_above_mean[-cnt_above:])
                df_run = pd.concat([df_run, df_above_mean[-cnt_above:]], ignore_index=True)
            if cnt_above>8:
                ab = cnt_above - 1
                df_run = df_run[:-ab]
                #df_run = df_run.append(df_above_mean[-cnt_above:])
                df_run = pd.concat([df_run, df_above_mean[-cnt_above:]], ignore_index=True)

        if idx>0 and val['Value']<=y_mean:
            #df_below_mean = df_below_mean.append(val)
            df_below_mean = pd.concat([df_below_mean, val], ignore_index=True)
            cnt_above = 0
            cnt_below += 1
            # check if below threshold
            if cnt_below==8:
                #df_run = df_run.append(df_below_mean[-cnt_below:])
                df_run = pd.concat([df_run, df_below_mean[-cnt_below:]], ignore_index=True)
            if cnt_below>8:
                bl = cnt_below - 1
                df_run = df_run[:-bl]
                #df_run = df_run.append(df_below_mean[-cnt_below:])
                df_run = pd.concat([df_run, df_below_mean[-cnt_below:]], ignore_index=True)

    return df_run
```

```python
def func_trend(DF: pd.DataFrame):
    df_trend = pd.DataFrame(columns=['Time', 'Value'])
    df_up_trend = pd.DataFrame(columns=['Time', 'Value'])
    df_down_trend = pd.DataFrame(columns=['Time', 'Value'])
    df_previous_value = pd.DataFrame(columns=['Time', 'Value'])
    cnt_up = 0
    cnt_down = 0

    for idx, val in DF.iterrows():
        # is this is the first value in the dataframe?
        if idx==0:
            df_previous_value = val

        # next value compare previous value
        if idx>0 and val['Value']>=df_previous_value['Value']:
            df_previous_value = val
            df_up_trend = df_up_trend.append(val)
            cnt_up += 1
            cnt_down = 0
            # check if above threshold
            if cnt_up==5:
                df_trend = df_trend.append(DF.iloc[idx-5])
                df_trend = df_trend.append(df_up_trend[-cnt_up:])
            if cnt_up>5:
                ab = cnt_up - 1
                df_trend = df_trend[:-ab]
                df_trend = df_trend.append(df_up_trend[-cnt_up:])

        if idx>0 and val['Value']<df_previous_value['Value']:
            df_previous_value = val
            df_down_trend = df_down_trend.append(val)
            cnt_up = 0
            cnt_down += 1
            # check if above threshold
            if cnt_down==5:
                df_trend = df_trend.append(DF.iloc[idx-5])
                df_trend = df_trend.append(df_down_trend[-cnt_down:])
            if cnt_down>5:
                bl = cnt_down - 1
                df_trend = df_trend[:-bl]
                df_trend = df_trend.append(df_down_trend[-cnt_down:])

    return df_trend
```

```python
df_trend = func_trend(df)
len(df_trend)
```

```python
def func_hug(DF: pd.DataFrame):
    df_zone = pd.DataFrame(columns=['Time', 'Value'])
    df_zone_c = pd.DataFrame(columns=['Time', 'Value'])
    df_not_zone_c = pd.DataFrame(columns=['Time', 'Value'])
    cnt_zone_c = 0

    y = DF['Value']
    y_mean = np.mean(y)
    std=DF['Value'].std()
    std_ucl_one=y_mean+(std*1)
    std_lcl_one=y_mean-(std*1)

    for idx, val in DF.iterrows():
        # is this is the first value in the dataframe?
        if idx==0:
            df_previous_value = val

        # next value compare previous value
        if idx>0 and std_lcl_one <= df_previous_value['Value'] <= std_ucl_one and std_lcl_one <= val['Value'] <= std_ucl_one:
            df_previous_value = val
            df_zone_c = df_zone_c.append(val)
            cnt_zone_c += 1
            # check if above threshold
            if cnt_zone_c==15:
                df_zone = df_zone.append(DF.iloc[idx-15])
                df_zone = df_zone.append(df_zone_c[-cnt_zone_c:])
            if cnt_zone_c>15:
                ab = cnt_zone_c - 1
                df_zone = df_zone[:-ab]
                df_zone = df_zone.append(df_zone_c[-cnt_zone_c:])
        else:
            df_previous_value = val
            cnt_zone_c = 0

    return df_zone
```

```python

def func_run(df: pd.DataFrame) -> pd.DataFrame:
    """
    Rule: 8 or more consecutive points all above or all below the mean.
    Returns subset of df with those points.
    """
    if df.empty:
        return df.iloc[0:0].copy()

    y = df['Value'].to_numpy()
    y_mean = y.mean()

    # +1 for > mean, -1 for <= mean (matches original logic)
    side = np.where(y > y_mean, 1, -1)

    # Group id whenever 'side' changes
    grp = np.zeros_like(side, dtype=int)
    grp[1:] = np.cumsum(side[1:] != side[:-1])

    tmp = pd.DataFrame({'grp': grp}, index=df.index)
    sizes = tmp.groupby('grp')['grp'].transform('size').to_numpy()

    mask = sizes >= 8
    return df.loc[mask].copy()


def func_trend(df: pd.DataFrame) -> pd.DataFrame:
    """
    Rule: 6 or more points in a row all increasing (>=) or all decreasing (<),
    measured as 5+ consecutive comparisons.
    Returns subset of df with those points.
    """
    n = len(df)
    if n < 2:
        return df.iloc[0:0].copy()

    v = df['Value'].to_numpy()
    diff = np.diff(v)  # length n-1, diff[i] = v[i+1] - v[i]

    up = diff >= 0    # non-decreasing
    down = diff < 0   # strictly decreasing
    mask = np.zeros(n, dtype=bool)

    def mark_runs(flag_arr: np.ndarray, min_len: int) -> None:
        """Mark runs of True in flag_arr with length >= min_len."""
        nonlocal mask
        if flag_arr.size == 0:
            return

        # Group id whenever flag flips True/False
        grp = np.zeros_like(flag_arr, dtype=int)
        grp[1:] = np.cumsum(flag_arr[1:] != flag_arr[:-1])
        idxs = np.arange(flag_arr.size)

        # Only groups where flag_arr is True
        true_groups = np.unique(grp[flag_arr])
        for g in true_groups:
            members = idxs[(grp == g) & flag_arr]
            run_len = members.size
            if run_len >= min_len:
                start = members[0]
                end = members[-1]
                # diff indices start..end correspond to values start..end+1
                mask[start:end + 2] = True  # mark those value indices

    # 5+ comparisons => at least 6 points
    mark_runs(up, 5)
    mark_runs(down, 5)

    return df.loc[mask].copy()


def func_hug(df: pd.DataFrame) -> pd.DataFrame:
    """
    Rule: 16 or more consecutive points hugging the mean within 1.
    (Original code counted 15 consecutive 'pairs' => 16 points.)
    Returns subset of df with those points.
    """
    if df.empty:
        return df.iloc[0:0].copy()

    y = df['Value']
    y_mean = y.mean()
    std = y.std()
    std_ucl_one = y_mean + std
    std_lcl_one = y_mean - std

    in_zone = (y >= std_lcl_one) & (y <= std_ucl_one)

    # bool_pair[i] is True when both i-1 and i are in zone
    bool_pair = in_zone & in_zone.shift(1, fill_value=False)
    arr = bool_pair.to_numpy()
    n = len(arr)
    mask = np.zeros(n, dtype=bool)

    if n == 0:
        return df.iloc[0:0].copy()

    grp = np.zeros_like(arr, dtype=int)
    grp[1:] = np.cumsum(arr[1:] != arr[:-1])
    idxs = np.arange(n)

    true_groups = np.unique(grp[arr])
    for g in true_groups:
        members = idxs[(grp == g) & arr]
        run_len = members.size
        # 15 consecutive 'pairs' -> 16+ hugging points
        if run_len >= 15:
            s = members[0]
            e = members[-1]
            # members s..e correspond to pairs (s-1, s)..(e-1, e)
            start_val = max(s - 1, 0)
            end_val = e
            mask[start_val:end_val + 1] = True

    return df.loc[mask].copy()


def func_4_5(df: pd.DataFrame) -> pd.DataFrame:
    """
    Rule: 4 out of 5 consecutive points are above +1 or below -1.
    This reproduces the original logic, including the way later windows
    'shrink' the tail of earlier clusters.
    """
    n = len(df)
    if n < 5:
        return df.iloc[0:0].copy()

    y = df['Value'].to_numpy()
    y_mean = y.mean()
    std = y.std()
    std_ucl_one = y_mean + std
    std_lcl_one = y_mean - std

    over = (y > std_ucl_one).astype(int)
    under = (y < std_lcl_one).astype(int)

    selected: list[int] = []

    for idx in range(n):
        if idx < 4:
            continue

        window_slice = slice(idx - 4, idx + 1)
        cnt_over = int(over[window_slice].sum())
        cnt_under = int(under[window_slice].sum())

        if cnt_over >= 4 or cnt_under >= 4:
            if selected:
                # Drop last 4 entries (matches df_4_5 = df_4_5[:-4])
                selected = selected[:-4]
            selected.extend(range(idx - 4, idx + 1))

    if not selected:
        return df.iloc[0:0].copy()

    # Remove duplicates while preserving order
    seen = set()
    unique_idx = []
    for i in selected:
        if i not in seen:
            seen.add(i)
            unique_idx.append(i)

    return df.iloc[unique_idx].copy()


def func_control_chart(df: pd.DataFrame):
    """
    Build and show the control chart with:
      - all data
      - outliers (1.5*IQR)
      - run rule (func_run)
      - trend rule (func_trend)
      - hugging rule (func_hug)
      - 4-of-5 rule (func_4_5)
    """
    y = df['Value']
    y_mean = y.mean()

    # OUTLIERS (IQR rule)
    q1 = y.quantile(0.25)
    q3 = y.quantile(0.75)
    iqr = q3 - q1
    lcl = q1 - (1.5 * iqr)
    ucl = q3 + (1.5 * iqr)
    btw_cl = y.between(lcl, ucl)
    df_outliers = df.loc[~btw_cl]

    # SIGMA ZONES
    std = y.std()
    std_ucl_one = y_mean + std
    std_ucl_two = y_mean + 2 * std
    std_ucl_three = y_mean + 3 * std
    std_lcl_one = y_mean - std
    std_lcl_two = y_mean - 2 * std
    std_lcl_three = y_mean - 3 * std

    # Apply rules with new implementations
    df_run = func_run(df)
    df_trend = func_trend(df)
    df_hug = func_hug(df)
    df_4_5 = func_4_5(df)

    # PLOT
    fig, ax = plt.subplots(1, figsize=(16, 6))
    ax.set_title('Control Chart', fontweight='bold', fontsize=26)
    ax.set_xlabel('Time', fontweight='bold')
    ax.set_ylabel('Values', fontweight='bold')

    ax.set_xlim(df['Time'].min(), df['Time'].max())
    ax.set_ylim((df['Value'] - 300).min(), (df['Value'] + 3000).max())

    # Sigma zones
    ax.axhspan(std_lcl_one, std_ucl_one, alpha=0.5)
    ax.axhspan(std_lcl_two, std_ucl_two, alpha=0.3)
    ax.axhspan(std_lcl_three, std_ucl_three, alpha=0.2)
    ax.axhline(y=y_mean, color='k')

    ax.xaxis.set_major_locator(mdates.DayLocator(interval=7))
    ax.xaxis.set_major_formatter(DateFormatter("%Y-%m-%d"))

    # All data
    ax.plot(df['Time'], df['Value'], linestyle='--', linewidth=0.6, marker='.', markersize=5, label='all data')
    # Outliers
    ax.plot(df_outliers['Time'], df_outliers['Value'], linestyle='none', marker='^', markersize=8, label='outliers')
    # Run
    ax.plot(df_run['Time'], df_run['Value'], linestyle='none', marker='o', markersize=7, label='run')
    # Trend
    ax.plot(df_trend['Time'], df_trend['Value'], linestyle='none', marker='>', markersize=5, label='trend')
    # Hugging
    ax.plot(df_hug['Time'], df_hug['Value'], linestyle='none', marker='P', markersize=7, label='hugging')
    # 4 out of 5
    ax.plot(df_4_5['Time'], df_4_5['Value'], linestyle='none', marker='D', markersize=5, label='4_5', markerfacecolor='red')

    ax.legend()
    fig.autofmt_xdate()
    plt.tight_layout()
    plt.show()


# Example call:
#func_control_chart(df.iloc[300:600])
#func_control_chart(df.iloc[600:900])
func_control_chart(df.iloc[750:1200])
```

```python
def waterfall(df, x, y):

    plt.rcParams["figure.figsize"] = (16,8)

    # calculate running totals
    df['tot'] = df[y].cumsum()
    df['tot1']=df['tot'].shift(1).fillna(0)


    # lower and upper points for the bar charts
    lower = df[['tot','tot1']].min(axis=1)
    upper = df[['tot','tot1']].max(axis=1)


    # mid-point for label position
    mid = (lower + upper)/2

    # positive number shows green, negative number shows red
    df.loc[df[y] >= 0, 'color'] = 'green'
    df.loc[df[y] < 0, 'color'] = 'red'

    # calculate connection points
    connect= df['tot1'].repeat(3).shift(-1)
    connect[1::3] = np.nan

    fig,ax = plt.subplots()

    # plot first bar with colors
    bars = ax.bar(x=df[x],height=upper, color =df['color'])

    # plot second bar - invisible
    plt.bar(x=df[x], height=lower,color='white')

    # plot connectors
    plt.plot(connect.index,connect.values, 'k' )

    # plot bar labels
    for i, v in enumerate(upper):
        plt.text(i-.15, mid[i], f"{df[y][i]:,.0f}")
```

```python
fig = go.Figure()

fig.add_trace(go.Barpolar(
    r=[77.5, 72.5, 70.0, 45.0, 22.5, 42.5, 40.0, 62.5],
    name='11-14 m/s',
    marker_color='rgb(106,81,163)'
))
fig.add_trace(go.Barpolar(
    r=[57.5, 50.0, 45.0, 35.0, 20.0, 22.5, 37.5, 55.0],
    name='8-11 m/s',
    marker_color='rgb(158,154,200)'
))
fig.add_trace(go.Barpolar(
    r=[40.0, 30.0, 30.0, 35.0, 7.5, 7.5, 32.5, 40.0],
    name='5-8 m/s',
    marker_color='rgb(203,201,226)'
))
fig.add_trace(go.Barpolar(
    r=[20.0, 7.5, 15.0, 22.5, 2.5, 2.5, 12.5, 22.5],
    name='< 5 m/s',
    marker_color='rgb(242,240,247)'
))

fig.update_traces(text=['North', 'N-E', 'East', 'S-E', 'South', 'S-W', 'West', 'N-W'])
fig.update_layout(
    title=dict(text='Wind Speed Distribution in Laurel, NE'),
    font_size=16,
    legend_font_size=16,
    polar_radialaxis_ticksuffix='%',
    polar_angularaxis_rotation=90,
    autosize=False,
    width=800,
    height=800,
)

fig.show()
```

```python

fig = go.Figure()


fig.add_trace(go.Barpolar(
    r=[77.5, 72.5, 70.0, 45.0, 22.5, 42.5, 40.0, 62.5],
    name='11-14 m/s',
    marker_color='rgb(106,81,163)',
    theta=['one','two','three','four','five','six','sev','eigh']
))
fig.add_trace(go.Barpolar(
    r=[57.5, 50.0, 45.0, 35.0, 20.0, 22.5, 37.5, 55.0],
    name='8-11 m/s',
    marker_color='rgb(158,154,200)',
    theta=['one','two','three','four','five','six','sev','eigh']
))
fig.add_trace(go.Barpolar(
    r=[40.0, 30.0, 30.0, 35.0, 7.5, 7.5, 32.5, 40.0],
    name='5-8 m/s',
    marker_color='rgb(203,201,226)',
    theta=['one','two','three','four','five','six','sev','eigh']
))
fig.add_trace(go.Barpolar(
    r=[20.0, 7.5, 15.0, 22.5, 2.5, 2.5, 12.5, 22.5],
    name='< 5 m/s',
    marker_color='rgb(242,240,247)',
    theta=['one','two','three','four','five','six','sev','eigh']
))


fig.update_traces(text=['11-14', '8-11', '5-8', '5', '5', '5', '5', '5'])
fig.update_layout(
    title=dict(text='My New Version'),
    font_size=16,
    legend_font_size=16,
    #polar_radialaxis_ticksuffix='%',
    polar_angularaxis_rotation=90,

    autosize=False,
    width=800,
    height=800,
    template="plotly_dark",
    #direction="counterclockwise",
    polar=dict(radialaxis=dict(tickvals = [None]))
)

fig.show()
```

```python
import plotly.express as px
df = px.data.wind()
fig = px.bar_polar(df, r="frequency", theta="direction",
                   color="strength", template="plotly_dark",
                   direction="counterclockwise",
                   color_discrete_sequence= px.colors.sequential.Plasma_r)

fig.update_layout(
    autosize=False,
    width=800,
    height=800,
)
fig.show()
```
